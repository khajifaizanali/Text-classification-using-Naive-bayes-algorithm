```python
#importing all data from files
import os
import numpy as np
all_files=os.listdir("C:/Users/HP/20_newsgroups")
print(all_files)
path="C:/Users/HP/20_newsgroups"
s='/'
i=0
dt=np.dtype('U25')
y=np.empty((19997),dtype=dt) #For sving output corresponding to each file available
filedata={}  # Will store data of files along with their location
for file in all_files:
    origin=path+s+file
    files=os.listdir(origin)
    for filename in files :
        y[i]=file
        i+=1
        name=origin+s+filename
        print(name)
        with open(name,"r") as f :
            filedata[name]=f.read()
```

```python
import nltk #importing natural language toolkit to download stopwords
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
en_stops = set(stopwords.words('english'))#all stopwords are stored in en_stops variable
#common words which are not relevant to document but are influencing in decreasing accuracy
commonwords=['Re:','In-Reply-To:','X-Copyright:','Nntp-Posting-Host:','Thanks,','Article-I.D.:','E-mail','Apr','GMT','wrote:','writes:','Lines:','Summary:','Organization:','Message-ID:','from','date',':', '>','->' ,'subject:', '**' , '/*' ,'1.1' , '28' , '0' , 'newsgroups', 'xref', 'path', 'from', 'subject', 'sender', 'organisation', 'apr','gmt', 'last','better','never','every','even','two','good','used','first','need','going','must','really','might','well','without','made','give','look','try','far','less','seem','new','make','many','way','since','using','take','help','thanks','send','free','may','see','much','want','find','would','one','like','get','use','also','could','say','us','go','please','said','set','got','sure','come','lot','seems','able','anything','put', '--', '|>', '>>', '93', 'xref', 'cantaloupe.srv.cs.cmu.edu', '20', '16', "max>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'ax>'", '21', '19', '10', '17', '24', 'reply-to:', 'thu', 'nntp-posting-host:', 're:','25''18'"i'd"'>i''22''fri,''23''>the','references:','xref:','sender:','writes:','1993','or']
```

```python
words={} #dictionary of words which are not in en_stop and commonwords list
#eaach word along with their frequency are stored in words dictionary
data=filedata.values()
for i in data:
    spi=i.split(" ")
    for j in spi:
        if j not in en_stops and j not in commonwords:
            if j in words.keys():
                words[j]+=1
            else:
                words[j]=1

    
```

```python
x=words.values()  #values of words are stored in x
z=words.keys()  #values of words are stored in z
```

```python
pwords={}
for i in z:
    if words[i]>400:  #words with frequency more than 400 are used in below algo because to decrease complexity and overfitting
        pwords[i]=words[i]  #words with more frequency are stored in pwords dictionary
```

```python
dt=[]
k=0
#creation of dataset
for i in data:  # this loop goes through entire data in all files
    spi=i.split(" ")  #data of a file is splitted into words and stored in 'spi'
    l=list(np.zeros(len(pwords.keys()))) #if a word in file is in pwords dictionary then respective index of word in 'l' list is incremented 
    for j in spi:
        if j in pwords.keys():
            ind=list(pwords.keys()).index(j)
            l[ind]+=1 
    dt.append(l)  
dt=np.array(dt)  
        
```

```python
def fit(x_train,y_train):  #fit function to fit the algo to respective dataset
    result={}
    a=set(y_train)
    for i in a:
        result[i]={}
        current_class=(y_train==i)  ##to distinguish a particular class from other classes
        xj=np.array(x_train)[current_class]
        yj=np.array(y_train)[current_class]
        fea=len(x_train[0])
        result["total_data"]=len(y_train)
        result[i]["total_count"]=0  #intialised to 0 to count total number of words
        result[i]["total_y"]=len(yj)
        for j in range(0,fea):
            result[i][j]=xj[:,j].sum() 
            result[i]["total_count"]+=xj[:,j].sum()
    return result
def predict(dic,xtest): #function used to classify test data
    y_pred=[]
    for i in xtest:
        y_pred.append(predict_single_point(dic,i))
    return y_pred
def predict_single_point(dic,i): #function is used to know the which class has more probability
    prob=-3000
    bestclass=-1
    for k in dic.keys():
        if (k == "total_data"):
            continue
        preal=probab(dic,i,k)
        if preal>prob:
            prob=preal
            bestclass=k
    return bestclass
def probab(dic,x,result): #we can know the probability based on dictionary ,words in a file and the expected classifier
    output=np.log(dic[result]["total_y"])-np.log(dic["total_data"])
    num_fea=len(dic[result].keys())-2
    for i in range(0,num_fea):
        if x[i]==0:
            continue
        numer=dic[result][i]+1 
        denom=dic[result]["total_count"]+len(x) #laplace is applied so we added lenght of x
        probab=np.log(numer)-np.log(denom)
        output=output+probab
    return output
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(dt,y,test_size=0.20,random_state=0) #dataset of words is splitted into training and testimg data
dictionary = fit(X_train,Y_train) #train dataset is fitted to algo
```

```python
Y_pred=predict(dictionary,X_test) #predicting the values of x_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix  #printing the classification_report and confusion matrix
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```
